```python
import pandas as pd
import numpy as np
```

### LEITURA DOS DADOS DOS 2 ARQUIVOS NECESSÁRIOS PARA O DESAFIO
#### Utilizando métodos para tratamento de dados direto da leitura dos arquivos

```python
df = pd.read_csv('prods_tab.csv',encoding='latin-1')
df_entregas = pd.read_csv('entregas_tab.txt', sep='|', skiprows=5, parse_dates=['DT.Compra', 'Dt.Chegada'])
```

```python
df_entregas
```

```python
df = df[(~df['ID. Forn.'].isna())] 
df = df[df['ID. Forn.']!='ID. Forn.']
df_entregas = df_entregas[(~df_entregas['SKTD'].isna())]
```

```python
df_entregas
```

### Removendo linhas e colunas irrelevantes para o processamento dos dados

```python
df
```

```python
ndf = df.drop(columns=['Prod. DESC.', 'GRP. MERC. 3', 'COD. MARC.'])
#ndf = ndf.dropna().drop(5)
#ndf_entregas = df_entregas.drop({0})
ndf_entregas = df_entregas.drop(columns=['Unnamed: 0', 'Unnamed: 5'])
```

```python
ndf
```

### Organizando dados .TXT

```python
ndf_entregas
```

```python
ndf_entregas['SKTD'] = ndf_entregas['Dt.Chegada'] - ndf_entregas['DT.Compra']
```

```python
ndf_entregas.columns
```

```python
ndf_entregas.rename(columns={'SKTD': 'Tempo_Entrega'}, inplace=True)
ndf_entregas = ndf_entregas.reset_index(drop = True)
```

```python
ndf_entregas
```

### Manipulando os dados .CSV

```python
ndf = ndf.reset_index(drop = True)
```

```python
ndf
```

### Trabalhando resultado final em um novo Data Frame

```python
ndf1 = pd.concat([ndf, ndf_entregas], axis=1)
```

```python
ndf1
```

```python
ndf1['Tempo_Entrega'] = ndf1['Tempo_Entrega'].astype(str)
ndf1['Dias'] =  ndf1['Tempo_Entrega'].apply(lambda x: x[:2])
ndf1['Dias'] = ndf1['Dias'].astype(float)
ndf1 = ndf1.drop(columns=['SKU.', 'DT.Compra', 'Dt.Chegada', 'Tempo_Entrega'])
```

```python
ndf1
```

```python
ndf1.columns
```

```python
ndf_finish = ndf1.groupby('ID. Forn.')['SKU.   '].count().reset_index()
```

```python
ndf_finish['lt'] = 0
```

```python
for i in range (4):
    ndf_finish.loc[0, 'lt'] = ndf1.loc[i, 'Dias'] + ndf_finish.loc[0, 'lt']
    
for i in range (6,8):
    ndf_finish.loc[1, 'lt'] = ndf1.loc[i, 'Dias'] + ndf_finish.loc[1, 'lt']
    
ndf_finish.loc[2, 'lt'] = ndf1.loc[4, 'Dias'] #+ ndf_finish.loc[2, 'lt']
ndf_finish.loc[3, 'lt'] = ndf1.loc[5, 'Dias'] + ndf1.loc[8, 'Dias']
    
#for i in range (7):
#    ndf_finish.loc[0, 'lt'] = ndf1.loc[i, 'Dias'] + ndf_finish.loc[0, 'lt']
#
#for i in range (1):
#    ndf_finish.loc[0, 'lt'] = ndf1.loc[i, 'Dias'] + ndf_finish.loc[0, 'lt']
    
    
```
